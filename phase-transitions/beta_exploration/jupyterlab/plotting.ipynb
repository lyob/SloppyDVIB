```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
filename = 'iterated-beta10.5.csv'
```

```python
epoch = []
IZY = []
IZX = []
acc = []

with open('../data-csv/{}'.format(filename)) as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        epoch.append(row[0])
        IZY.append(row[1])
        IZX.append(row[2])
        acc.append(row[3])
        
def convert_to_floats(array):
    out = [float(element) for element in array]
    return out

epoch = convert_to_floats(epoch)
acc = convert_to_floats(acc)
IZY = convert_to_floats(IZY)
IZX = convert_to_floats(IZX)
```

```python
# print(epoch)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(epoch, acc)
ax.set_xlabel('epoch')
ax.set_ylabel('acc')
ax.set_title('accuracy vs epoch')
_ = plt.xticks(range(0, 1000, 100))
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(IZX, IZY)
ax.set_xlabel('IZX')
ax.set_ylabel('IZY')
ax.set_title('IZY vs IZX by epoch')
# _ = plt.xticks()
```
