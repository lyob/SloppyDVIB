```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
all_eps = []
all_izy = []
all_izx = []
all_acc = []

for i in range(25):
    beta = i / 2
    
    filename = 'iterated-beta{}.csv'.format(beta)
    
    epoch = []
    IZY = []
    IZX = []
    acc = []

    with open('data-csv/{}'.format(filename)) as csvDataFile:
        csvReader = csv.reader(csvDataFile)
        for row in csvReader:
            epoch.append(row[0])
            IZY.append(row[1])
            IZX.append(row[2])
            acc.append(row[3])

    def convert_to_floats(array):
        out = [float(element) for element in array]
        return out

    epoch = convert_to_floats(epoch)
    acc = float(acc[-1])
    IZY = float(IZY[-1])
    IZX = float(IZX[-1])
    
    all_eps.append(epoch)
    all_izx.append(IZX)
    all_izy.append(IZY)
    all_acc.append(acc)
```

```python
print(all_izy)
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax.scatter(all_izx[:], all_izy[:], marker='.')
ax.set_xlabel('izx')
ax.set_ylabel('izy')
ax.set_title('I(Z;Y) vs I(Z;X) as a function of beta')
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax.scatter(np.linspace(0, 12, 25), all_izx[:], marker = '.')
ax.set_xlabel('beta (1e-x)')
ax.set_ylabel('I(Z;X)')
ax.set_title('I(Z;X) vs beta')
_ = plt.xticks(np.linspace(0, 12, 13))
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax.scatter(np.linspace(0, 12, 25), all_acc, marker = '.')
ax.set_xlabel('beta (1e-x)')
ax.set_ylabel('acc')
ax.set_title('accuracy vs beta')
_ = plt.xticks(np.linspace(0, 12, 13))
```
